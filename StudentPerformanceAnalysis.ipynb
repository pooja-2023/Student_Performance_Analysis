```python
import pandas as pd
```

```python
student=pd.read_csv('student.csv')
```

```python
student
```

```python
score=pd.read_csv('score.csv')
```

```python
score
```

```python
merged_data=pd.merge(student,score,on='student_id')
```

```python
merged_data
```

```python
aggregated_data=merged_data.groupby(['student_id','name']).agg({
    'score':'mean'
}).reset_index()
    
```

```python
aggregated_data
```

```python
min_score= aggregated_data['score'].min()
```

```python
min_score
```

```python
max_score= aggregated_data['score'].max()
```

```python
max_score
```

```python
aggregated_data['normalised_score']=(aggregated_data['score']-min_score)/(max_score-min_score)
```

```python
normalised_score
```

```python
aggregated_data
```

```python
aggregated_data['score_category']=pd.cut(
    aggregated_data['normalised_score'],
    bins=3,
    labels=['Lowest','Medium','Highest']
)
```

```python
aggregated_data
```

```python
merged_data=pd.concat([merged_data ,merged_data.iloc[:]])
```

```python
merged_data.to_csv("merged_data.csv", index=False)
```
