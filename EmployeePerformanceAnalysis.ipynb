```python
import pandas as pd
```

```python
employee=pd.read_csv('employee.csv')
```

```python
employee
```

```python
performance=pd.read_csv('performance.csv')
```

```python
performance
```

```python
merged_data=pd.merge(employee,performance,on='employee_id')
```

```python
merged_data
```

```python
aggregated_data=merged_data.groupby(['employee_id','name']).agg({
    'task_score':'mean'
}).reset_index()
```

```python
aggregated_data
```

```python
min_score= aggregated_data['task_score'].min()
```

```python
min_score
```

```python
max_score= aggregated_data['task_score'].max()
```

```python
max_score
```

```python
aggregated_data['normalised_score']=(aggregated_data['task_score']-min_score)/(max_score-min_score)
```

```python
aggregated_data
```

```python
aggregated_data['score_category']=pd.cut(
    aggregated_data['normalised_score'],
    bins=3,
    labels=['Lowest','Medium','Highest']
)
```

```python
aggregated_data
```

```python
merged_data=pd.concat([merged_data ,merged_data.iloc[:]])
```

```python
merged_data.to_csv("merged_data.csv", index=False)
```
